```python
## Commments
## Case sensitivity     -- Python is case sensitive

name = "Yogesh"
Name = "Pharswan"
print(name)
print(Name)
```

```python
## Indentation -- use to define a block of code in Python. It is used to indicate which statements belong to which blocks of code.Consistent indentation is crucial in Python, as it determines the structure and flow of the program. Typically, a standard indentation level is 4 spaces, but you can use any number of spaces as long as you are consistent throughout your code.
age = 25
if age >= 18:
    print("You are an adult.")
else:    print("You are a minor.")
```

```python
##Line continuation -- In Python, you can use a backslash (\) to indicate that a statement continues on the next line. This is useful for improving readability when you have long lines of code.
total = 10 + 20 + 30 + \
3+0 + 40+ 50
print(total)
```
