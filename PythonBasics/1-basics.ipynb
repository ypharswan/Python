```python
## Naming conventions -- In Python, there are several naming conventions that are commonly followed to improve code readability and maintainability. Some of the most common naming conventions include:
# 1. Snake Case: This convention uses lowercase letters and underscores to separate words in variable and function names. For example: my_variable, calculate_sum(), etc.
# 2. Camel Case: This convention capitalizes the first letter of each word except the first one. It is often used for class names. For example: MyClass, EmployeeDetails, etc.
# 3. Pascal Case: This convention capitalizes the first letter of each word, including the first one. It is also used for class names. For example: MyClass, EmployeeDetails, etc.
# 4. Constants: Constants are typically written in uppercase letters with underscores separating words. For example: MAX_VALUE, PI, etc.

first_name = "John"  # Snake Case
LastName = "Doe"     # Camel Case
print(first_name)
print(LastName) 
```

```python
## Python is a dynamically typed language, which means that you do not need to declare the type of a variable when you create it. The type of a variable is determined at runtime based on the value assigned to it. This allows for greater flexibility in coding, as you can easily change the type of a variable by assigning a different value to it. For example:
age = 25  # age is an integer
print(type(age))  # Output: <class 'int'>
name = "Yogesh"  # name is a string
print(type(name))  # Output: <class 'str'>
```

### Arithmatic Operators

```python
## Arithmatic Operations

a = 10
b = 20
add_result = a + b # addition
print("Addition:", add_result)
sub_result = a - b # subtraction
print("Subtraction:", sub_result)
mult_result = a * b # multiplication
print("Multiplication:", mult_result)
division_result = a / b # division
print("Division:", division_result)
floor_division_result = a // b # floor division
print("Floor Division:", floor_division_result)
modulus_result = a % b # modulus
print("Modulus:", modulus_result)
exponential_result = a ** b # exponential
print("exponential_result",exponential_result)
```
