Example 1  - Manage a To Do List

```python
to_do_list = ["Clean the house", "Buy groceries", "Pay bills"]

## Adding to tasks
to_do_list.append("Call mom")
to_do_list.append("Finish project")
to_do_list.append("Exercise")

## Removing completed tasks
to_do_list.remove("Clean the house")

if "Pay bills" in to_do_list:
    print("Don't forget to pay the bills!")

print("To do list remaining")
for task in to_do_list:
    print(f'-{task}')
```

Example 2 - Organizing Students Grades

```python
grades = [85, 92, 67,78, 90, 88]
grades.append(95)  # Adding a new grade
grades.append(82)  # Adding another grade

## Calculating average grade
average_grade = sum(grades) / len(grades)
print(f'Average grade: {average_grade:.2f}')

max_grade = max(grades)
min_grade = min(grades)
print(f'Max grade: {max_grade}')
print(f'Min grade: {min_grade}')
```

Example 3 - Managing Inventory
