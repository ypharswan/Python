### Tuples - Tuples are ordered collection of items that are immutable. hey are similiar to list but their immutability makes them different.

```python
## Creating tuple

empty_tuple = ()
print(empty_tuple)
```

```python
lst = list()
print(lst)  
print(type(lst))
tup = tuple()
print(tup)
print(type(tup))
```

```python
numbers = tuple([1, 2, 3, 4, 5])
numbers
```

```python
list1 =list((1,2,3,4,5))
list1
```

```python
mixed_tuple = (1, "Hello", 3.14, [1, 2, 3], (4, 5))
mixed_tuple
```

```python
## Accessing tuple elements
numbers
```

```python
print(numbers[0])
print(numbers[-1])
print(numbers[1:4])
print(numbers[::-1])
```

```python
## Tuple operations
concatination_tuple = numbers + mixed_tuple
print(concatination_tuple)
```

```python
mixed_tuple * 3
```

```python
## Immutability of tuples
## Tuples are immutable means elements cannot be changed once assigned

lst = [1,2,6,3,4,5]
lst[0] = "Yogesh"
print(lst)

number = (1,2,4,5,6)
# number[0] = "Yogesh"              ## 'tuple' object does not support item assignment
# print(number)
```

```python
## Tuple Methods
print(numbers)

print(numbers.count(1))
print(numbers.index(5))
```

```python
str_tup = ("my","name","is","khan","and","I","am","not","and","terrorist")
print(str_tup)
print(str_tup.count("and"))
print(str_tup.index("and"))
```

```python
## Packing and Unpacking of tuples
## Packing of tuple

packed_tuple = 1,"Hello",True
print(packed_tuple)
```

```python
## Unpacking of tuple
a,b,c = packed_tuple
print(a)
print(b)
print(c)
```

```python
## Unpacking with *
numbers =(1,2,3,4,5,6,7,8)
first,*middle, last = numbers
print(first)
print(middle)
print(last)
```

```python
## Nested List
lst = [[1,2,3,4],[5,6,7,8],[9,0,"Namaste",True]]
print(lst[1])
print(lst[1][3])
print(lst[0][0:3])
```

```python
lst = [[1,2,3,4],[5,6,7,8],(9,0,"Namaste",True)]
print(lst[1])
print(lst[2][3])
print(lst[2][0:3])
```

```python
nested_tuple = ((1,2,3),(9,0,"Namaste",True),(3,5,6),(True,False))
print(nested_tuple[1][2])
```

```python
## Iterating over nested tuple
for sub_tup in nested_tuple:
    for item in sub_tup:
        print(item, end = " ")
    print()
```

### Tuples are versatile and useful in many real world scenario where an immutable and ordered collectionm of items are required. They are commonly used in data structure , function argument and return values and as dictionary keys. Understanding how to  leverage tuples effectively can improve the efficiency and readability of your python code.
