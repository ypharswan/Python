### Functions in Python 
A function is a block of code that perform a specific  task. Functions help in organising code, reuse code and improve readabilty.

```python
## Syntax
def function_name(parameter):
    """"Docstring """
    # Function body
    return expression
```

```python
def even_or_odd(num):
    """" This function tells if the number is odd or even"""
    if num % 2 == 0:
        print("The number is even")
    else:
        print("The number is odd")
```

```python
## Calling the function
even_or_odd(79)
```

```python
## Function with multiple parameter

def Add(a,b):
    """"Add two number"""
    return a + b
result = Add(7,9)
print(result)
```

```python
## Default parameters
def greet(name = "Guest "):
    print(f"Hello {name} Welcome to the Paradise")
greet("Yogesh")
```

```python
greet()
```

```python
### Variable Length Arguments
## Positional  and keyword arguments


## Poaitional arguments
def print_numbers(*args):
    for num in args:
        print(num)
```

```python
print_numbers(1,2,3,"Hello",4,6,4,79,)
```

```python
## Keyword Arguments
def print_details(**kwargs):
    for key,val in kwargs.items():
        print(f"{key}-{val}")
print_details(name = "Rohit",age =45,country ="India",Runs = 98989)
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f"Positional Argument - {val}")
    for key,val in kwargs.items():
        print(f"{key}-{val}")
```

```python
## Positional argument cannot appear after keyword arguments
print_details(1,2, 4,6,7,"Yogesh",name = "Sachin",Run = 9878987, Age =54)
```
